```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid', font_scale = 1.5)
# load data
correlations = pd.read_csv('./output/vars_sorted.csv')
```

```python
ax.set_ylim(0, 1)
ax.figure
```
