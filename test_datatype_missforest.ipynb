```python
import pandas as pd
import numpy as np

# need to know categorical variables
from configs import PreprocessingConfig

# missforest
from msap.modeling.model_selection.preprocessing import Preprocessor
from missingpy import MissForest

# list of variables
vars_cat_prefs = PreprocessingConfig.columns_categorical
# not encoded
unencoded = pd.read_csv("./output/data_cleaned.csv")
# encoded
encoded = pd.read_csv("./output/data_cleaned_encoded_18_yesmental.csv")
```

```python
encoded.dtypes
```

```python
for col in encoded.columns:
    if any([col.startswith(x) for x in vars_cat_prefs]):
        encoded[col] = encoded[col].astype("category")

```

```python
encoded.dtypes
```

```python
def get_all_categorical(
        categorical: list,
        X: pd.DataFrame):

    names = []
    for feature in X.columns:
        if feature.startswith(tuple(categorical)):
            names.append(feature)
    # print(names)
    indices = [X.columns.get_loc(c) for c in names if c in X]
    return indices
```

```python
cat_vars = get_all_categorical(vars_cat_prefs, encoded)
```

```python
imputer = MissForest(n_jobs=1, verbose=0, random_state=42)
encoded = encoded.head(10)
print(encoded)
encoded.drop('y18CH_Dep_YN_216m', axis=1)
print(encoded)
data_new_df = pd.DataFrame(
            imputer.fit_transform(X=encoded, cat_vars=cat_vars),
            index=encoded.index,
            columns=encoded.columns)
```
